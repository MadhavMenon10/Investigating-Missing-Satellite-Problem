## Imports

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import scipy as scp
```

## Creating transfer function and MPK

```python
def transfer_function(P_x1, P_x2):
    transfer_function_list = []
    for i in range(len(P_x1)):
        transfer_function_list.append(np.sqrt(P_x1[i]/P_x2[i]))
    return transfer_function_list
```

# $\Lambda$CDM and $\Lambda$WDM

## MPK For $\Lambda$CDM

```python
plt.rcParams.update({'font.size': 12})
data_cdm = np.loadtxt("cdm_2022stdnu00_pk.dat")
df_cdm = pd.DataFrame(data_cdm)
k_cdm = list(df_cdm[0])
P_cdm = list(df_cdm[1])
plt.plot(k_cdm, P_cdm)
plt.xscale("log")
plt.yscale("log")
plt.xlim(1e-4,6)
plt.ylim(1,5e4)
plt.xlabel(r"k (h Mpc$^{-1}$)")
plt.ylabel(r"P(k) [(Mpc h$^{-1}$)$^3$]")
plt.title("Matter Power Spectrum for ΛCDM")
plt.savefig('MPK for CDM',bbox_inches='tight')
plt.show()
```

```python
data_wdm = np.loadtxt("wdm(neutrino)_2022stdnu01_pk.dat")
df = pd.DataFrame(data_wdm)
k_wdm = list(df[0])
P_wdm = list(df[1])
plt.plot(k_wdm, P_wdm)
plt.xscale("log")
plt.yscale("log")
plt.xlim(1e-4,6)
plt.ylim(1,5e4)
plt.xlabel(r"k (h Mpc$^{-1}$)")
plt.ylabel(r"P(k) [(Mpc h$^{-1}$)$^3$]")
plt.title("Matter Power Spectrum for ΛWDM")
plt.savefig('MPK for WDM',bbox_inches='tight')
plt.show()
```

```python
transfer_function_cdm_wdm = transfer_function(P_wdm, P_cdm)
plt.xscale("log")
plt.xlim(1e-1, 100)
plt.plot(k_cdm, transfer_function_cdm_wdm)
plt.title("Relative Transfer Function for $\Lambda$CDM and $\Lambda$WDM")
plt.xlabel(r"k (h Mpc$^{-1}$)")
plt.ylabel(r"T(k)")
plt.savefig("TF for CDM and WDM")
plt.show()
```

## Fitting functions

```python
k = np.array(list(df_g3[0]))
g = np.array([1.5, 3.0, 4.5, 6.0])
x, y = np.meshgrid(k, g)
x[x==0]=10**-5
y[y==0]=10**-5
z = np.zeros(x.shape)
z[0,:] = transfer_function_cdm_wdm
z[1,:] = transfer_function_cdm_g3
z[2,:] = transfer_function_cdm_g4point5
z[3,:] = transfer_function_cdm_g6
print(z)
```

```python
def _transthermal_fit(xy, *args): # You should use the different values of g here
    k, g = xy
    alpha, kappa, nu = args
    h = 0.6736
    Omega = 0.12*(h**-2)
    mu = alpha*(g/1.5)**kappa*((4*Omega)**0.11)*(((10*h)/7)**1.22)*h**-1
    transfer_function = (1+(mu*k)**(2*nu))**(-5/nu)
    return transfer_function

p0 = [0.049, -1.0, 1.12] # alpha guess is 0.049, kappa guess is -1, nu guess is 1.12

x_data = np.vstack((x.ravel(), y.ravel()))
potb, pcov = scp.optimize.curve_fit(_transthermal_fit, x_data, z.ravel(), p0)
print("potb:", potb) # The function finds the values for alpha, kappa, and nu
print("pcov:", pcov)
```
