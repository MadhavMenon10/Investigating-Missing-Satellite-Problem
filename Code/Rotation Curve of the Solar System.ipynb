## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

## Plotting figure 1

```python
df = pd.read_csv("Planetary data - Sheet1.csv")
df
```

```python
r = []
v = []
for planet in range(1, 10):
    if planet!=4:
        r.append(float(df[df.columns[planet]][7]))
        v.append(float(df[df.columns[planet]][11]))

plt.plot(r, v,marker='o')
plt.title("Rotation curve of the solar system")
plt.xlabel("R $(10^6$ km)")
plt.ylabel("v (kms$^-1$)")
plt.annotate("Mercury",(145, 46.25))
plt.annotate("Venus",(200, 34))
plt.annotate("Earth",(250, 29))
plt.annotate("Mars",(300, 23.6))
plt.annotate("Jupiter",(925, 14))
plt.annotate("Saturn",(1500, 10))
plt.annotate("Uranus",(2950, 8))
plt.annotate("Neptune",(3950, 7))
plt.savefig('RotationCurveSolarSystem.jpg',bbox_inches='tight')
plt.show()
```
